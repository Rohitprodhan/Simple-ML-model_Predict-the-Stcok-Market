```python
import yfinance as yf
import pandas as pd
data=yf.download("^GSPC")
```

```python
data.to_csv("sp500.csv")
```

```python
data.index
```

```python
data.plot.line(y="Close",use_index=True)
```

```python
sp500=data.loc["1990-01-01":].copy()
```

```python
sp500.plot.line(y="Close",use_index=True)
```

```python
sp500["Tommorow"]=sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"]=(sp500["Tommorow"]>sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500.dropna(axis=0)
```

```python
# Inital Model 
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100, min_samples_split=100,random_state=1)

train=sp500.iloc[:-100]
test=sp500.iloc[-100:]

predictors= ["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
pred=model.predict(test[predictors])
```

```python
# Validation
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(test["Target"],pred)
```

```python
MAE
```

```python
# Model Upgradation
# Adding new Columns

sp500["Daily_Return"] = sp500["Close"].pct_change() #Fractional change between the current and a prior element.
# sp500['MA_7'] = sp500['Close'].rolling(window=7).mean()
# sp500['MA_30'] = sp500['Close'].rolling(window=30).mean()

sp500['Return_Volatility'] = sp500['Daily_Return'].rolling(window=20).std()


```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+= [ratio_column, trend_column]
```

```python
sp500.dropna(inplace=True)
```

```python
train=sp500.iloc[:-100].copy()
test=sp500.iloc[-100:].copy()

```

```python
train
```

```python
predictors= ["Close","Volume","Open","High","Low","Daily_Return","Return_Volatility"]
model.fit(train[predictors],train["Target"])
```

```python
pred=model.predict(test[predictors])
```

```python
# Validation
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(test["Target"],pred)
```

```python
MAE
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
precision_score(test["Target"], preds)
```
